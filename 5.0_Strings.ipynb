### Behaviour of slice operator:
- s[start:stop:step]
- step value can be either +ve or ve
- if +ve then it should be forward direction(left to right) and we have to consider bEgin to end-1
- if -ve then it should be backward direction(right to left) and we have to consider bEgin to end+1

<b>***Note: In the backward direction if end value is -1 then result is always empty. In the forward direction if end value is 0 then result is always empty.</b>

#### In forward direction:
- default value for bEgin: 0 
- default value for end: length of string 
- default value for step: +1

#### In backward direction:
- default value for bEgin: -1 
- default value for end: -(length of string+1)

<b>Note: Either forward or backward direction, we can take both +ve and -ve values for bEgin and end index</b>

#### Q. Write a program to access each character of string in forward and backward direction by using while loop?

```python
s = "hi good morning"
n = len(s) 
i = 0

print("Forward direction")
while i<n:
    print(s[i],end='')
    i=i+1
print()    
print("Backward direction")
i = -1
while i>=-n:
    print(s[i],end='')
    i=i-1
```

```python
s = "hi good morning"
print("Forward direction")
for i in s[::]:
    print(i, end='')
print()
    
print("Backward direction")
for i in s[::-1]:
    print(i, end='')
```

## 6. Finding subStrings
- find()   -> Returns index of first occurrence of the given substring. If it is not available then we will get -1
- index()  -> it is same as find() method except that if the specified substring is not available then we will get ValueError.
- rfind()  -> returns last index value if not returns -1
- rindex() -> returns last index value if not returns ValueError

```python
s="cat ter pillar cat"

print("****find method****")
print(s.find("cat"))
print(s.find("pillar"))
print(s.find("dog"))

print("****index method****")
print(s.index("cat"))
#print(s.index("dog"))      # Value Error: substring not found

print("****rfind method****")
print(s.rfind("r"))
print(s.rfind("cat"))

print("****rindex method****")
print(s.rindex("cat"))
#print(s.rindex("dog"))  # Value Error: substring not found
```

```python
s = input("enter the main string:")
sub = input("enter the sub string:")

try:
    s.index(sub)
except ValueError:
    print("sub string not found.")
else:
    print("sub string found.")
```

<b>Note: By default find() method can search total string. We can also specify the boundaries to search.</b>
- syntax: s.find(subs,start,stop)
- It will always search from bEgin index to end-1 index

```python
s="cat ter pillar cat"
print(len(s))
print(s.find("cat",0,17))
print(s.find("cat",3,17))
```

#### Q. Program to display all positions of substring in a given main string

```python
s = "sapabapa"
sub = "a"
flag = False
pos = -1
n = len(s)

while True:
    pos = s.find(sub, pos+1, n)
    if pos == -1:
        break
    print("Found at position:",pos)
    flag = True
if flag == False:
    print("Not Found")
```

## 9. Splitting of String
- We can split the given string according to specified seperator by using split() method.
- s.split(separator)
- The default seperator is space. The return type of split() method is List

```python
s="pavan kumar puppala"
l = s.split()
print(l)

for i in l:
    print(i)
```

```python
d="09-12-1994"
lst = d.split("-")
print(lst)

for i in lst:
    print(i)
```
